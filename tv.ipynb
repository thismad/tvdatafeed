```python
from tvDatafeed import TvDatafeed,Interval
from dotenv import load_dotenv
import os
import plotly.express as px
load_dotenv()
```

```python
# initialize tradingview
tv = TvDatafeed(username=os.getenv('TV_USERNAME'),password=os.getenv('TV_PASSWORD'),auth_token=os.getenv('TV_AUTH_TOKEN'))
```

```python
df = tv.get_hist('BTCUSDT','BINANCE',Interval.in_1_hour,n_bars=30000)
df.reset_index(inplace=True)
df.sort_values('datetime',inplace=True)
px.line(df,x='datetime',y='close').show()
```

```python
nifty_data=tv.get_hist('NIFTY','NSE',interval=Interval.in_1_hour,n_bars=1000)
px.line(nifty_data,x='time',y='close').show()
```

```python
nifty_data.info()
```

```python
nifty_data.close.plot(figsize=(15,10),use_index=False)
```
